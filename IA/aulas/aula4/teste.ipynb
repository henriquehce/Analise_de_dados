```python
import os
from dotenv import load_dotenv

# 1) tenta carregar .env local
load_dotenv()

# 2) tenta ler secret do Colab (se estiver no Colab)
api_key = None
try:
    from google.colab import userdata
    api_key = userdata.get("gemini") or userdata.get("GOOGLE_API_KEY")
except Exception:
    pass

# 3) fallback para variáveis de ambiente locais
api_key = (
    api_key
    or os.getenv("GOOGLE_API_KEY")
    or os.getenv("GEMINI_API_KEY")
    or os.getenv("API_KEY")
    or os.getenv("chave")
)

if not api_key:
    raise ValueError(
        "Chave não encontrada. Defina GOOGLE_API_KEY no .env ou nas variáveis de ambiente."
    )

os.environ["GOOGLE_API_KEY"] = api_key
os.environ["chave"] = api_key

print("Chave API carregada com sucesso.")
```
