```python
# Exercício 1: Divisão de dois números float com tratamento de erro
try:
    num1 = float(input("Digite o primeiro número: "))
    num2 = float(input("Digite o segundo número: "))
    resultado = num1 / num2
    print(f"O resultado da divisão é: {resultado}")
except ValueError:
    print("Erro: ValueError - Um ou ambos os valores digitados não são números válidos.")
except ZeroDivisionError:
    print("Erro: ZeroDivisionError - Não é possível dividir por zero.")

# Exercício 2: Busca em dicionário com tratamento de KeyError
idades = {'Júlia': 16, 'Carol': 23, 'Alberto': 19, 'Roberta': 17}
try:
    chave = input("Digite o nome para buscar a idade: ")
    valor = idades[chave]
    print(f"A idade de {chave} é {valor}")
except KeyError:
    print("Nome não encontrado")

# Exercício 3: Converter lista para float
def converter_para_float(lista):
    try:
        lista_convertida = [float(item) for item in lista]
        return lista_convertida
    except ValueError as e:
        print(f"Erro: ValueError - {e}")
        return None
    except TypeError as e:
        print(f"Erro: TypeError - {e}")
        return None
    finally:
        print("Fim da execução da função")

# Exercício 4: Agregar listas com tratamento de erro
def agregar_listas(lista1, lista2):
    try:
        if len(lista1) != len(lista2):
            raise IndexError("A quantidade de elementos em cada lista é diferente.")
        resultado = [(lista1[i], lista2[i], lista1[i] + lista2[i]) for i in range(len(lista1))]
        return resultado
    except TypeError as e:
        print(f"Erro: TypeError - Valores incompatíveis para soma: {e}")
        return None
    except IndexError as e:
        print(f"Erro: IndexError - {e}")
        return None

# Exercício 5: Contabilizar pontuações
def calcular_notas(gabarito, testes):
    notas = []
    for teste in testes:
        for alternativa in teste:
            if alternativa not in ['A', 'B', 'C', 'D']:
                raise ValueError(f"A alternativa {alternativa} não é uma opção de alternativa válida")
        nota = sum([1 for i in range(len(gabarito)) if teste[i] == gabarito[i]])
        notas.append(nota)
    return notas

gabarito = ['D', 'A', 'B', 'C', 'A']
testes_sem_ex = [['D', 'A', 'B', 'C', 'A'], ['C', 'A', 'A', 'C', 'A'], ['D', 'B', 'A', 'C', 'A']]
testes_com_ex = [['D', 'A', 'B', 'C', 'A'], ['C', 'A', 'A', 'E', 'A'], ['D', 'B', 'A', 'C', 'A']]

try:
    print("Notas sem exceção:", calcular_notas(gabarito, testes_sem_ex))
except ValueError as e:
    print(e)

try:
    print("Notas com exceção:", calcular_notas(gabarito, testes_com_ex))
except ValueError as e:
    print(e)

# Exercício 6: Verificar pontuações em palavras
def verificar_pontuacao(lista_palavras):
    for palavra in lista_palavras:
        if ',' in palavra or '.' in palavra or '!' in palavra or '?' in palavra:
            raise ValueError(f'O texto apresenta pontuações na palavra "{palavra}".')
    print("Texto validado com sucesso!")

lista_tratada = ['Python', 'é', 'uma', 'linguagem', 'de', 'programação', 'poderosa', 'versátil',
                  'e', 'fácil', 'de', 'aprender', 'utilizada', 'em', 'diversos', 'campos', 'desde',
                  'análise', 'de', 'dados', 'até', 'inteligência', 'artificial']

lista_nao_tratada = ['Python', 'é', 'uma', 'linguagem', 'de', 'programação', 'poderosa,', 'versátil',
                  'e', 'fácil,', 'de', 'aprender', 'utilizada', 'em', 'diversos', 'campos,', 'desde',
                  'análise', 'de', 'dados', 'até', 'inteligência', 'artificial!']

try:
    verificar_pontuacao(lista_tratada)
except ValueError as e:
    print(e)

try:
    verificar_pontuacao(lista_nao_tratada)
except ValueError as e:
    print(e)

# Exercício 7: Dividir colunas com tratamento de erro
def divide_colunas(pressoes, temperaturas):
    try:
        if len(pressoes) != len(temperaturas):
            raise ValueError("As listas têm tamanhos diferentes.")
        
        resultado = []
        for i in range(len(pressoes)):
            if temperaturas[i] == 0:
                raise ZeroDivisionError(f"Divisão por zero detectada na posição {i}.")
            resultado.append(pressoes[i] / temperaturas[i])
        
        return resultado
    except ValueError as e:
        print(f"Erro: ValueError - {e}")
        return None
    except ZeroDivisionError as e:
        print(f"Erro: ZeroDivisionError - {e}")
        return None

# Teste sem exceção
pressoes = [100, 120, 140, 160, 180]
temperaturas = [20, 25, 30, 35, 40]
print("Resultado sem exceção:", divide_colunas(pressoes, temperaturas))

# Teste com ZeroDivisionError
pressoes = [60, 120, 140, 160, 180]
temperaturas = [0, 25, 30, 35, 40]
print("Resultado com ZeroDivisionError:", divide_colunas(pressoes, temperaturas))

# Teste com ValueError
pressoes = [100, 120, 140, 160]
temperaturas = [20, 25, 30, 35, 40]
print("Resultado com ValueError:", divide_colunas(pressoes, temperaturas))
```
