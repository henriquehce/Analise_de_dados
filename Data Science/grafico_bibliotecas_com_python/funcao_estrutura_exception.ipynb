```python
#notas dos estudantes por semestre

notas = {'1° trimenstre': 85,
         '2° trimenstre': 90,
         '3° trimenstre': 78,
         '4° trimenstre': 88}

soma_notas = sum(notas.values()) / len(notas)
print(f'A média anual dos estudantes foi: {soma_notas}')

#ou

notas_2 = {'1° trimenstre': 85,
           '2° trimenstre': 60,
           '3° trimenstre': 75,
           '4° trimenstre': 99}

soma = 0
for nota in notas_2.values():
    soma += nota
media = soma / len(notas_2)
print(f'A média anual dos estudantes foi, em outro método: {media}')
```

```python
#notas aleatorias com random
from random import randint
notas_3 = {'1° trimenstre': randint(0, 100),
           '2° trimenstre': randint(0, 100),
           '3° trimenstre': randint(0, 100),
           '4° trimenstre': randint(0, 100)}

def calcular_media_anual(notas_3):
    soma = 0
    for nota in notas_3.values():
        soma += nota
    media = soma / len(notas_3)
    return media

def situacao_final(media):
    if media >= 60:
        return "Aprovado"
    else:
        return "Reprovado"

print(notas_3)
media_anual = calcular_media_anual(notas_3)
print(f'A média anual dos estudantes foi: {media_anual}, logo ele está {situacao_final(media_anual)}.')
```

```python
#função de media com list

def media_list(lista: list) -> float:
    soma = sum(lista)
    return soma / len(lista)
print(media_list([85, 90, 78, 88, 92]))
```
